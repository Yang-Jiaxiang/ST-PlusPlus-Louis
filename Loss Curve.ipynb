```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def readCSV(path):
    df = pd.read_csv(path)
    return df
```

```python
def plotCurve(df):
    plt.figure(figsize=(15, 10))

    # 創建 2x2 的圖表佈局
    fig, axs = plt.subplots(2, 2, figsize=(15, 10))

    # 第一張圖 - 監督損失
    axs[0, 0].plot(df['t_loss'], label='Training Loss')
    axs[0, 0].plot(df['v_loss'], label='Validation Loss')
    axs[0, 0].set_xlabel('Epoch')
    axs[0, 0].set_ylabel('Loss')
    axs[0, 0].set_title('Training and Validation Loss Curve')
    axs[0, 0].legend()
    axs[0, 0].grid(True)
    axs[0, 0].set_ylim(0, 1)
    
    # 第三張圖 - 一致性損失
    axs[0, 1].plot(df['t_miou'], label='Training miou')
    axs[0, 1].plot(df['v_miou'], label='Validation miou')
    axs[0, 1].set_xlabel('Epoch')
    axs[0, 1].set_ylabel('Miou')
    axs[0, 1].set_title('Training and Validation Miou Curve')
    axs[0, 1].legend()
    axs[0, 1].grid(True)
    axs[0, 1].set_ylim(0, 1)

    # 第四張圖 - 準確率
    axs[1, 0].plot(df['t_dice'], label='Training dice')
    axs[1, 0].plot(df['v_dice'], label='Validation dice')
    axs[1, 0].set_xlabel('Epoch')
    axs[1, 0].set_ylabel('Dice')
    axs[1, 0].set_title('Training and Validation Dice Curve')
    axs[1, 0].legend()
    axs[1, 0].grid(True)
    # 調整局，防止標題和軸標籤重疊
    plt.tight_layout()
    axs[1, 1].set_ylim(0, 1)

    # 顯示圖表
    plt.show()
```

# ST

## supervised

```python
csv_file_path = 'outdir/loss/loss_supervised_labeled.csv'# 請將此處替換為你的 CSV 檔案路徑
df = readCSV(csv_file_path)
plotCurve(df)
```
